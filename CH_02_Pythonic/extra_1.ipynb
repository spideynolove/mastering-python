```python
# Define a list of numbers
numbers = [1, 2, 3, 4, 5]

# Use functools.reduce() to calculate the product of all elements
product = reduce(lambda x, y: x * y, numbers)

# Print the result
print("Product of all numbers:", product)
```

```python
# Define the number of Fibonacci numbers to generate
n = 10

# Use functools.reduce() to calculate Fibonacci numbers
fibonacci_series = reduce(lambda x, _: x + [x[-2] + x[-1]], range(n - 2), [0, 1])

# # Add the first two numbers manually since they are not generated by reduce
# fibonacci_series = [0, 1] + fibonacci_series

# Print the Fibonacci series
"Fibonacci Series:", fibonacci_series
```

```python
from collections import ChainMap

# Define dictionaries with default values
default_config = {'debug': False, 'verbose': False}
user_config = {'debug': True}       # overrides default

# Create a ChainMap with user_config taking precedence
config = ChainMap(user_config, default_config)

# Access config values
print(config['debug'])     # True (from user_config)
print(config['verbose'])   # False (from default_config)
```

```python
from collections import ChainMap

# Define different context dictionaries
global_config = {'user': 'admin', 'level': 'user'}
local_config = {'level': 'admin'}

# Create a ChainMap to represent the current context
current_context = ChainMap(local_config, global_config)
current_context['user'], current_context['level']
```

```python
# Switch to a different context
new_context = {'level': 'manager'}
current_context = current_context.new_child(new_context)    # overrides level

# Access the current context
current_context['user'], current_context['level']
```

```python
from collections import ChainMap

current_context = ChainMap({}, {})


def create_config(current_context=None, do_switch=True):
    # Define different context dictionaries
    global_config = {'user': 'admin', 'level': 'user'}
    local_config = {'level': 'admin'}

    # Create a ChainMap to represent the current context
    current_context = ChainMap(local_config, global_config)

    def switch_context(new_context):
        nonlocal current_context
        current_context = current_context.new_child(new_context)

    # Switch to a different context
    switch_context({'level': 'manager'}) if do_switch else None
    return current_context


current_context = create_config(current_context, True)
current_context['user'], current_context['level']
```
